```python
import os
import ollama
import json
from io import StringIO
```

```python

# Query notebook: this notebook runs the same prompt across multiple LLMs and stores the responses.
# Create new entry for each LLM you want to run against. See https://ollama.com/library for a list of supported models.
llm_test_list = ['qwen:1.8b','qwen:4b','deepseek-r1:1.5b','gemma:2b','qwen2.5:1.5b','qwen2.5:3b','phi3', 'starcoder2', 'llama3.2:3b','phi']
llm_test_list = ["granite3.1-dense:2b","granite3.1-moe","exaone3.5:2.4b", "tinyllama:1.1b-chat-v1-q8_0","tinyllama"]
llm_responses = []

prompt = 'What would the vikings do if they discovered a smartphone? Ensure your response is cogent and follows logically, consistent with the real-world capabilities of your average Viking.'

# Iterate through each LLM and ask it the same question, and print to console. 
for llm in llm_test_list:
    ollama.pull(llm)  # Pull model if it has not already been pulled. There will be a slight delay while the model pulls for the first time
    response = ollama.chat(model=llm, messages=[
        {
            'role': 'user',
            'content': prompt,
        },
    ])
    
    response_content = response['message']['content']
    llm_responses.append((llm, response_content))
    
    # Write to file and print to screen.
    with open("outputfile.txt", "a") as f:
        f.write(f"{llm} response:\n{response_content}\nEnd response.\n----------------\n")
    
    print(f"{llm}\nresponse:\n{response_content}\nEnd response.\n----------------\n")
```
